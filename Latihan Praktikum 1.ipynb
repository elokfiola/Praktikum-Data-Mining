### Data Prepocessing

```python
#Identifikasi nilai yang hilang (Missing Values)
df.isnull().sum()
```

```python
#Identifikasi Duplikat Data
df.groupby('BMI')["BMI"].agg("count")
```

```python
#Identifikasi Outlier
correlation_matrix = df.corr() #menghitung korelasi antar kolom
sns.heatmap(correlation_matrix)
print(correlation_matrix)
plt.show()
sns.boxplot(x='BMI', data=df)
```

```python
#Identifikasi data tidak berimbang (Data Imbalanced)
df['Outcome'].unique() #Menampilkan variabel pada atribut Outcome
df['Outcome'].value_counts() #Menampilkan komposisi pada atribut Outcome
```

```python
df['Outcome'].unique()
#Misal 0 untuk orang yang tidak terjangkit diabetes dan 1 untuk orang ang terjangkit diabetes
```
