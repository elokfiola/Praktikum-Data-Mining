```python
#import library
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Menentukan 10 saham
grab = 'GRAB'
swn = 'SWN'
oxy = 'OXY'
futu = 'FUTU'
meta = 'META'
nke = 'NKE'
path = 'PATH'
gap = 'GAP'
ebay = 'EBAY'
ggb = 'GGB'
```

```python
#tentukan nama data dan hari pertama dan hari terakhir nya
a = yf.download(grab, start="2023-09-1", end="2024-09-1")
b = yf.download(swn, start="2023-09-1", end="2024-09-1")
c = yf.download(oxy, start="2023-09-1", end="2024-09-1")
d = yf.download(futu, start="2023-09-1", end="2024-09-1")
e = yf.download(meta, start="2023-09-1", end="2024-09-1")
f = yf.download(nke, start="2023-09-1", end="2024-09-1")
g = yf.download(path, start="2023-09-1", end="2024-09-1")
h = yf.download(gap, start="2023-09-1", end="2024-09-1")
i = yf.download(ebay, start="2023-09-1", end="2024-09-1")
j = yf.download(ggb, start="2023-09-1", end="2024-09-1")
```

```python
#Memberi batas pengambilan data saham dengan close
grab_close = a[['Close']]
swm_close = b[['Close']]
oxy_close = c[['Close']]
futu_close = d[['Close']]
meta_close = e[['Close']]
nke_close = a[['Close']]
path_close = b[['Close']]
gap_close = c[['Close']]
ebay_close = d[['Close']]
ggb_close = e[['Close']]
grab_close.rename(columns={'Close': 'GRAB'}, inplace=True)
swm_close.rename(columns={'Close': 'SWM'}, inplace=True)
oxy_close.rename(columns={'Close': 'OXY'}, inplace=True)
futu_close.rename(columns={'Close': 'FUTU'}, inplace=True)
meta_close.rename(columns={'Close': 'META'}, inplace=True)
nke_close.rename(columns={'Close': 'NKE'}, inplace=True)
path_close.rename(columns={'Close': 'PATH'}, inplace=True)
gap_close.rename(columns={'Close': 'GAP'}, inplace=True)
ebay_close.rename(columns={'Close': 'EBAY'}, inplace=True)
ggb_close.rename(columns={'Close': 'GGB'}, inplace=True)
```

```python
#Menggabungkan data saham menggunakan fungsi concat
data_saham = pd.concat([grab_close, swm_close, oxy_close, futu_close, meta_close, nke_close, path_close, gap_close, ebay_close, ggb_close], axis = 1)
print(data_saham)
```

```python
#melihat data kosong/ missing value

#membuat data frame
df = pd.DataFrame(data_saham)
df.isna().sum()
```

Ternyata semua data saham tidak ada missing value/ artinya semua dat alengkap dan dapat dilanjutkan ke tahap analisis lanjutan

## Normalisasi data

```python
#Membuat grafik harga penutupan data saham sebelum dinormalisasi
plt.figure(figsize=(12, 6))
plt.plot(data_saham['GRAB'], label='Saham GRAB', color='red')
plt.plot(data_saham['SWM'], label='Saham SWM', color='yellow')
plt.plot(data_saham['OXY'], label='Saham OXY', color='blue')
plt.plot(data_saham['FUTU'], label='Saham FUTU', color='green')
plt.plot(data_saham['META'], label='Saham META', color='pink')
plt.plot(data_saham['NKE'], label='Saham NKE', color='lightblue')
plt.plot(data_saham['PATH'], label='Saham PATH', color='orange')
plt.plot(data_saham['GAP'], label='Saham GAP', color='purple')
plt.plot(data_saham['EBAY'], label='Saham EBAY', color='maroon')
plt.plot(data_saham['GGB'], label='Saham GGB', color='magenta')

#Menambahkan judul dan label
plt.title(f'Grafik Harga Saham {data_saham}')
plt.xlabel('Tanggal')
plt.ylabel('Harga Saham')
plt.legend()
plt.grid()

#Menampilkan grafik
plt.show()
```

```python
#Normalisasi data dengan MinMax

import pandas as pd
df_normalized=(df - df.min()) / (df.max() - df.min())
print("Data Normalized (Min-Max):\n", df_normalized)
```

```python
#Normalisasi data dengan MinMax sklearn

from sklearn.preprocessing import MinMaxScaler
#Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

#Normalisasi menggunakan Scikit-learn
norm_minmax = pd.DataFrame (scaler.fit_transform(data_saham), columns=data_saham.columns)
print("Data Normalized (Min-Max) dengan Scikit-learn: \n", norm_minmax)
```

```python
#normalisasi data dengan Z score

#Menghitung rata-rata (mean) dan deviasi standar (standard deviation)
mean = df.mean()
std_dev = df.std()

#Menghitung Z-score menggunakan rumus
df_zscore1 = (df - mean) / std_dev
print("Data Normalized (Z-Score): \n", df_zscore1)
```

```python
#normalisasi data dengan Z score sklearn

#Inisialisasi StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#Melakukan standarisasi
norm_z = pd.DataFrame (scaler.fit_transform(data_saham), columns=data_saham.columns)
print("Data Normalized (Z-Score) dengan Scikit-learn: \n", norm_z)
```

```python
#normalisasi dat adengan deccimal scaling

import pandas as pd
#Contoh DataFrame
df1=pd.DataFrame (data_saham)

#Menentukan nilai maksimum
max_value = df1.max().max()
#Menentukan j (jumlah digit dari nilai maksimum)
j = len(str(max_value))
#Melakukan normalisasi dengan decimal scaling
norm_desimal = df1 / (10** j)
print("Data Normalized (Decimal Scaling): \n", norm_desimal)
```
