### Forward Selection



```python
#Import Library

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.metrics import accuracy_score

#Import data dan membaca data
dfsepeda = pd.read_csv('bike-sharing.csv')

#menampilkan data teratas
dfsepeda.head()
```

```python
#menghapus missing value
dfsepeda = dfsepeda.dropna()
```

```python
#Conversi data tanggal
dfsepeda['dteday'] = pd.to_datetime(dfsepeda['dteday'])
dfsepeda['day'] = dfsepeda['dteday'].dt.day
dfsepeda['month'] = dfsepeda['dteday'].dt.month
dfsepeda['year'] = dfsepeda['dteday'].dt.year

#menantukan fitur (x) dna target (y)
X = dfsepeda[['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit', 'temp', 'atemp','hum', 'windspeed', 'casual', 'registered' ]]
y = dfsepeda['cnt']

#membagi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

# Model yang digunakan
model = RandomForestClassifier(random_state=42)

# Seleksi Maju (Forward Selection)
sfs = SequentialFeatureSelector(model, n_features_to_select=2, direction='forward')
sfs.fit(X_train, y_train)

print("Fitur terpilih:", sfs.get_support(indices=True))

#Transformasi Data
X_train_selected = sfs.transform(X_train)
X_test_selected = sfs.transform(X_test)

#Pelatihan Model dan Prediksi
model.fit(X_train_selected, y_train)
y_pred = model.predict(X_test_selected)

#Menghitung Akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model dengan fitur yang dipilih:", accuracy)
```

```python
#Menggunakan lebih sedikit fitur untuk mempercepat komputasi
#menantukan fitur (x) dna target (y)
X = dfsepeda[['holiday', 'weekday','windspeed' ]]
y = dfsepeda['cnt']

#membagi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

# Model yang digunakan
model = RandomForestClassifier(random_state=42)

# Seleksi Maju (Forward Selection)
sfs = SequentialFeatureSelector(model, n_features_to_select=2, direction='forward')
sfs.fit(X_train, y_train)

print("Fitur terpilih:", sfs.get_support(indices=True))

#Transformasi Data
X_train_selected = sfs.transform(X_train)
X_test_selected = sfs.transform(X_test)

#Pelatihan Model dan Prediksi
model.fit(X_train_selected, y_train)
y_pred = model.predict(X_test_selected)

#Menghitung Akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model dengan fitur yang dipilih:", accuracy)
```

Menggunakan 3 fitur: holiday', 'weekday','windspeed' dan menargetkan y = cnt. Menghasilkan fitur terpilihnya adalalh indeks 0 (holiday) dan 1 (windspeed). Dengan akurasi sangat kecil yaitu 1%. Hal ini

```python
#Menggunakan  fitur lain untuk menemukan fungsi target yang lebih tinggi
#menantukan fitur (x) dna target (y)
X = dfsepeda[['weathersit', 'atemp','windspeed' ]]
y = dfsepeda['registered']

#membagi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

# Model yang digunakan
model = RandomForestClassifier(random_state=42)

# Seleksi Maju (Forward Selection)
sfs = SequentialFeatureSelector(model, n_features_to_select=2, direction='forward')
sfs.fit(X_train, y_train)

print("Fitur terpilih:", sfs.get_support(indices=True))

#Transformasi Data
X_train_selected = sfs.transform(X_train)
X_test_selected = sfs.transform(X_test)

#Pelatihan Model dan Prediksi
model.fit(X_train_selected, y_train)
y_pred = model.predict(X_test_selected)

#Menghitung Akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model dengan fitur yang dipilih:", accuracy)
```

Menggunakan 3 fitur yaitu weathersit', 'atemp','windspeed'fitur terpilihnya pada indeks 0 dan 1 yaitu weathersit dan atempt. Dengan kedua fitur tersebut menghasilkan akurasi sekitar 0.014 atau hanya 1% pada data pengujian menghasilkan model sangat baik dalam mengklasifikasikan jadwal sepedah berdasarkan fitur yang dipilih.   

Akurasi yang kecil mungkin disebabkan oleh pemilihan fitur yang kurang tepat atau metode yang kurang tepat dengan fitur yang dipilih

## Backward selection

```python
# Import library
from sklearn.feature_selection import SequentialFeatureSelector

#Backward Selection
sbs = SequentialFeatureSelector(model, n_features_to_select=2, direction='backward')
sbs.fit(X, y)

# Fitur yang dipilih
print("Fitur yang paling relevan:", sbs.get_support(indices=True))
X_selected = sbs.transform(X) #Transformasi Data
model.fit(X_selected, y) #Pelatihan Model
```

Hal ini menunjukkan bahwa fitur yang terpilih adalah fitur dengan indeks 0 (weathersit) dan 2 (windspeed). Berarti kedua model memilih indeks 0 dan 2 sebagai fitur yang paling relevan

## Bidirectional selection

```python
#seleksi dua arah
sfs_bidirectional = SequentialFeatureSelector(model, n_features_to_select=2,

#pemberian toleransi pemberhentian
direction='forward', tol=1e-4)

sfs_bidirectional.fit(X, y)

#Mencetak fitur yang relevan untuk dipilih
print("Fitur yang paling releavn:", sfs_bidirectional.get_support(indices=True))

# Evaluasi model dengan fitur
X_selected = sfs_bidirectional.transform(X)
model.fit(X_selected, y)
```

Hal ini menunjukkan bahwa Bidirectional selection memilih fitur dengan indeks 0 (weathersit) dan 2 (windspeed). Berarti kedua model memilih indeks 0 dan 2 sebagai fitur yang paling relevan

## Metode Random Generation

```python
#Import library
import numpy as np
from itertools import combinations #menentukan kombinasi fitur
from sklearn.metrics import accuracy_score #untuk menghitung akurasi

# Pilih fitur secara acak
jumlah_fitur = X.shape[1]
kombinasie = list(combinations(range(jumlah_fitur), 2)) # Kombinasi 2 fitur
akurasi_terbaik = 0 #inisialisasi kombinasi awal = 0
kombinasi_terbaik = None #inisialisasi awal tidak ada kombinasi terbaik

for comb in kombinasie: #melakukan percabangan menggunakan for
  # Use .iloc to select columns by integer index
  X_selected = X.iloc[:, list(comb)] #pemilihan x
  model.fit(X_selected, y) #latih model menggunakan x yang terpilih demgan target (y)
  y_pred = model.predict(X_selected) #prediksi y
  acc = accuracy_score(y, y_pred) #menghitung akurasi pada kelas target fitur yang terpilih
  if acc > akurasi_terbaik: #melakuka pemilihan berdasarkan kondisi
    akurasi_terbaik = acc #perumpamaan fungsi akurasi menjadi variabel baru
    kombinasi_terbaik = comb #perumpamaan fungsi kombinasi menjadi variabel baru
print(f"Kombinasi terbaik: {kombinasi_terbaik}")
print(f"Akurasi:{akurasi_terbaik}")
```

Hal ini menunjukkan bahwa Random Generation memilih kombinasi (1,2) yaitu "weathersit dan atempt. Berarti kedua model memilih indeks 1 dan 2 sebagai kombinasi fitur yang paling relevan. Akurasi sangat tinggi dibandingkan dengan metode lainnya yaitu 0.08. Yang artinya fitur yang say apilih lebih cocok menggunakna metode random generator dengan akurasi 8%
